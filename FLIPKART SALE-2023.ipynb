### What is the distribution of ratings across different brands?

```python
# Calculate the distribution of ratings across different brands
rating_distribution = df.groupby('Brand')['Rating'].value_counts(normalize=True).unstack().fillna(0)

# Get the top 10 brands based on the total number of ratings
top_10_brands = df['Brand'].value_counts().nlargest(10).index
rating_distribution_top_10 = rating_distribution.loc[top_10_brands]

# Visualize the distribution of ratings across the top 10 brands
plt.figure(figsize=(14, 8))
rating_distribution_top_10.plot(kind='bar', stacked=True, colormap='viridis', figsize=(14, 8))
plt.title('Distribution of Ratings Across Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Proportion of Ratings')
plt.legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Insight
- Apple and Samsung stand out for having the highest proportions of ratings in the upper ranges (e.g., 4.5 to 4.8). Apple's ratings are particularly concentrated in the higher range, suggesting a strong positive customer sentiment.
- Redmi and Vivo also show strong performance in higher ratings but with more variation than Apple and Samsung.
- Kechaoda has a more varied distribution, with noticeable proportions of lower ratings (e.g., 3.0 to 4.0). This indicates a mixed customer experience or satisfaction level.
- Itel and Infinix similarly have significant portions of mid-range ratings, suggesting that while some customers are satisfied, many are less impressed.

### Insights:

- Price Plateaus: In some cases, the price may plateau or increase less steeply beyond a certain RAM size, indicating diminishing returns on higher RAM configurations.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,root_mean_squared_error
```

```python
# Example: Creating new features or aggregating existing ones
df['price_to_rating_ratio'] = df['Price'] / (df['Rating'] + 1)  # Example feature

# Selecting features and target variable
features = df.drop('Price', axis=1)  
target = df['Price']

# Split data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,root_mean_squared_error
```

```python
# Preprocessing for numerical and categorical data
numeric_features = [ 'Rating','RAM','ROM','Battery','Camera', 'price_to_rating_ratio'] 
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])
```

```python
categorical_features = ['Brand', 'Name']  # Example categorical features
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor

import pickle
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])
```

```python
# Build the model pipeline
model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('regressor', RandomForestRegressor())])
```

```python
# Train the model
model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(x_test)

# Evaluate the model
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R^2 Score:", r2_score(y_test, y_pred))
print("rmse:", root_mean_squared_error (y_test, y_pred))
print("mean_absolute_error:",mean_absolute_error(y_test, y_pred))
```

```python

y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
r2_train = r2_score(y_train, y_pred_train)
r2_test = r2_score(y_test, y_pred_test)

accuracy_metrics = {'algorithm': RandomForestRegressor.__name__,
                    'R2_train' : r2_train,
                    'R2_test'  : r2_test}

```

```python
accuracy_metrics 
```

```python
def machine_learning_regression(df, algorithm):

    features = df.drop('Price', axis=1)  
    target = df['Price']

    # Split data into train and test sets
    x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

    model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('regressor', algorithm())])
    model.fit(x_train, y_train)
    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)
    r2_train = r2_score(y_train, y_pred_train)
    r2_test = r2_score(y_test, y_pred_test)

    # algo = str(algorithm).split("'")[1].split(".")[-1]
    accuracy_metrics = {'algorithm': algorithm.__name__,
                        'R2_train' : r2_train,
                        'R2_test'  : r2_test}

    return accuracy_metrics
```

```python
print(machine_learning_regression(df, DecisionTreeRegressor))
print(machine_learning_regression(df, ExtraTreesRegressor))
print(machine_learning_regression(df, RandomForestRegressor))
print(machine_learning_regression(df, AdaBoostRegressor))
print(machine_learning_regression(df, GradientBoostingRegressor))
```
